```python
books = pd.read_csv('/kaggle/input/7k-books-with-metadata/books.csv')
```

```python
books = books[['isbn10','title','authors','categories','description','published_year','num_pages','ratings_count']]
```

```python
print(books['description'])
```

```python
books = books.astype(str)
```

```python
books['tags'] = books['description'] + books['authors'] + books['categories'] + books['published_year'] + books['num_pages'] + books['ratings_count'] 
```

```python
new = books.drop(columns=['description','authors','categories','published_year','num_pages','ratings_count'])

```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(book):
    index = new[new['title'] == book].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Spider\'s Web')
```

```python
recommend('The One Tree')
```
